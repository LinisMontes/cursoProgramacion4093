```python
num = int(input("Digite el numero: "))

if num > 20:
  print("El numero es mayor a 20")
elif num > 15:
  print("El numero es mayor a 15")
elif num > 10:
  print("El numero es mayor a 10")
elif num > 5:
  print("El numero es mayor a 5")
else:
  print("No se cumple ninguna condicion")
```

```python
cal1 = float(input("Digite la calificacion 1: "))
cal2 = float(input("Digite la calificacion 2: "))
cal3 = float(input("Digite la calificacion 3: "))

prom = round((cal1 + cal2 + cal3)/3, 2)

if prom < 3.0:
  estado = "Reprobado"
else:
  estado = "Aprobado"
print(f"El promedio es= {prom} y el estado es {estado}")
```
