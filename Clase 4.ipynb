Escribir un programa que realice la tabla de multiplicar de un numero digitado por teclado

```python
num = int(input("Digite el numero de la tabla de multiplicar: "))

for i in range(11):
  print(i * num)
```

```python
num = int(input("Digite el numero de la tabla de multiplicar: "))

for i in range(11):
  print(i, "x", num, "=", i * num)
```

Escriba un programa que sume los numeros pares del 0 al 10 y saque el total de la suma

```python
# variables ACUMULADORAS
acu = 0

for i in range(0,12,2):
  #print(i)
  acu = acu + i
print(f"El total de la suma de los numeros pares del 0-10 = {acu}")
```
